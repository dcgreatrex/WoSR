## Set up Environment and Get Data

```python
# To support both python 2 and python 3
from __future__ import division, print_function, unicode_literals

# Import packages
import numpy as np
import pandas as pd
import os
import urllib
import tarfile

# to ensure 'randomisation' is the same every time 
np.random.seed(5)

# Set up plotting
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12

# data source
train_url = "http://d396qusza40orc.cloudfront.net/predmachlearn/pml-training.csv"
test_url = "http://d396qusza40orc.cloudfront.net/predmachlearn/pml-testing.csv"
# destination file names
train_file = "AZDS_train.csv"
test_file = "AZDS_test.csv"
```

### Load data

```python
# download data, if not already
if not os.path.isfile(train_file):
    urllib.request.urlretrieve(train_url, train_file)
    print('downloading file: {0}'.format(train_file))
if not os.path.isfile(test_file):    
    urllib.request.urlretrieve(test_url, test_file)
    print('downloading file: {0}'.format(test_file))

# load data
training_set = pd.read_csv(train_file)
test_set = pd.read_csv(test_file)
```

#### separate features and label (outcome)

```python
strat_train_Y = pd.DataFrame(strat_train_set["classe"])
strat_train_X = strat_train_set.drop("classe", axis=1) 
```

#### scale

```python
# minmax and not standard, since many are not normally distributed. If there are outliers then use robustscaler
from sklearn.preprocessing import MinMaxScaler

indices = strat_train_X.index.values    # keep indices
features = strat_train_X.columns.values # keep feature names
scaler = MinMaxScaler()
strat_train_X = pd.DataFrame(scaler.fit_transform(strat_train_X)) # scaler returns array, not dataframe
strat_train_X.columns = features
strat_train_X.index = indices
```

#### remove features with 0 variance

```python
#https://www.programcreek.com/python/example/93973/sklearn.feature_selection.VarianceThreshold
#https://csiu.github.io/blog/update/2017/03/06/day10.html
# R is quite more automatic for this: https://www.r-bloggers.com/near-zero-variance-predictors-should-we-remove-them/
# but that also could be dangerous - we don't want to throw away low var data
from sklearn.feature_selection import VarianceThreshold

indices = strat_train_X.index.values    # keep indices
features = strat_train_X.columns.values # keep feature names
selector = VarianceThreshold()
strat_train_X = pd.DataFrame(selector.fit_transform(strat_train_X))
strat_train_X.columns = features
strat_train_X.index = indices
#selector.get_support(indices=True)
```

```python
# check
sns.boxplot(x="classe", y="pitch_forearm", data=pd.concat([strat_train_X, strat_train_Y], axis=1));
```

### Prepare the (hold-out) validation set to be used for model evaluation

```python
# stratified split
strat_valid_Y = pd.DataFrame(strat_valid_set["classe"])
strat_valid_X = strat_valid_set.drop("classe", axis=1) 

# scale
indices = strat_valid_X.index.values    # keep indices
features = strat_valid_X.columns.values # keep feature names
scaler = MinMaxScaler()
strat_valid_X = pd.DataFrame(scaler.fit_transform(strat_valid_X)) # scaler returns array, not dataframe
strat_valid_X.columns = features
strat_valid_X.index = indices

# remove features with 0 variance
indices = strat_valid_X.index.values    # keep indices
features = strat_valid_X.columns.values # keep feature names
selector = VarianceThreshold()
strat_valid_X = pd.DataFrame(selector.fit_transform(strat_valid_X))
strat_valid_X.columns = features
strat_valid_X.index = indices

# label encoding
indices = strat_valid_Y.index.values    # keep indices
features = strat_valid_Y.columns.values # keep feature names
lb = LabelEncoder()
strat_valid_Ycat = pd.DataFrame(lb.fit_transform(strat_valid_Y))
strat_valid_Ycat.columns = features
strat_valid_Ycat.index = indices
```

### Logistic regression

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(random_state=2, solver='lbfgs')
log_reg.fit(strat_train_X, strat_train_Ycat)
```

```python
predictions = log_reg.predict(strat_train_X)
```

```python
from sklearn.metrics import confusion_matrix
conf = confusion_matrix(strat_train_Ycat, predictions)
```

```python
import itertools

classes = np.unique(strat_train_Y)
conf_norm = conf.astype('float') / conf.sum(axis=1)[:,np.newaxis]                                    # plot normalized matrix
plt.imshow(conf_norm, cmap=plt.cm.Blues, interpolation='None')
plt.colorbar()
plt.grid(False)
tick_marks = np.arange(len(classes))
plt.xticks(tick_marks, classes)
plt.yticks(tick_marks, classes)
fmt = '.2f' 
thresh = conf_norm.max() / 2.
for i, j in itertools.product(range(conf_norm.shape[0]), range(conf_norm.shape[1])):
    plt.text(j, i, format(conf_norm[i, j], fmt),
             horizontalalignment="center",
             color="white" if conf_norm[i, j] > thresh else "black")

#plt.tight_layout()
plt.ylabel('True label')
plt.xlabel('Predicted label')

plt.show()
# false positives are row-wise. 
# false negatives are column-wise.
# precision (tp/(tp+fp)) i.e. not to label something as positive when in reality it is negative
# recall (tp/(tp+fn)), i.e. to find all the positives (diagonal in matrix is the per-class recall)
```

```python
conf_norm
```

```python
from sklearn.metrics import precision_recall_fscore_support

# this shows precision, recall, f-score, and the number of occurences in each class
#precision_recall_fscore_support(strat_train_Ycat, predictions, beta=1.0, labels=None, 
#                                pos_label=1, average=None, warn_for=('precision', 'recall', 'f-score'), sample_weight=None)

# precision, recall, f-score, and support (number of real trues - no sense here)
precision_recall_fscore_support(strat_train_Ycat, predictions, average='weighted')
```
